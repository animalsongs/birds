## Libraries

Import the required classes and functions from `wavesongs`:

```python
%matplotlib ipympl

from wavesongs.objects.song import Song
from wavesongs.utils.paths import ProjDirs
from wavesongs.utils import plots
from wavesongs.utils.tools import get_measures
```

```python
proj_dirs = ProjDirs(audios="./assets/audio", results="./assets/results")

tlim_roi = (0.7, 2.7)

copeton_song = Song(proj_dirs, file_id="574179401", tlim=tlim_roi)
copeton_song.acoustical_features(umbral_FF=1.4, NN=256, ff_method="yin", flim=(1e2, 2e4))

plots.spectrogram_waveform(copeton_song, tlim=tlim_roi, save=False)
```

```python
copeton_song.play()
```

```python
klicker_data = plots.spectrogram_data(copeton_song)
```

```python
get_measures(klicker_data, copeton_song, save=False)
```
